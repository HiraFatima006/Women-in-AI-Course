```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("hb.png",cv2.IMREAD_COLOR)
cv2.imshow("image",img)   
cv2.waitKey(0)
```

```python
half=cv2.resize(img,(0,0),fx=0.5,fy=0.5)
cv2.imshow("half",img)
cv2.waitKey(0)
```

```python
enlarge=cv2.resize(img,(1000,1000))
cv2.imshow("enlarge",img)
cv2.waitKey(0)
```

```python
stretch_near=cv2.resize(img,(780,540),interpolation=cv2.INTER_NEAREST)
cv2.imshow("stretch_near",img)
cv2.waitKey(0)
```

```python
Titles=['Original','Half','Enlarged','Stretched']
images=[img,half,enlarge,stretch_near]
for i in range(4):
  plt.subplot(2,3,i+1)
  plt.title(Titles[i])
  plt.imshow(images[i])
```

```python
# Gaussian Blur
```

```python
Gaussian_Blur=cv2.GaussianBlur(img,(7,7),0)
cv2.imshow("Gaussian Blurring",Gaussian_Blur)
cv2.waitKey(0)
```

```python
# Median Blur
```

```python
Median_Blur=cv2.medianBlur(img,5)
cv2.imshow("Median Blurring",Median_Blur)
cv2.waitKey(0)
```

```python
# Bilateral Blur
```

```python
Bilateral_Blur=cv2.bilateralFilter(img,9,75,75)
cv2.imshow("Bilateral Blurring",Bilateral_Blur)
cv2.waitKey(0)
```

```python
Edged=cv2.Canny(img,500,500)
cv2.imshow("edge",Edged)
cv2.waitKey(0)
```

```python
Title=['Original','Half','Enlarged','Stretched','Gaussian Blur','Median Blur','Bilateral Blur','Edged']
images=[img,half,enlarge,stretch_near,Gaussian_Blur, Median_Blur, Bilateral_Blur, Edged]
for i in range(8):
  plt.subplot(2,4,i+1)
  plt.title(Title[i])
  plt.imshow(images[i])
```

```python
(rows,cols)=img.shape[:2]
print(rows,cols)
M=cv2.getRotationMatrix2D((cols/2,rows/2),45,1)
res=cv2.warpAffine(img,M,(200,200))
cv2.imshow("res",res)
cv2.waitKey(0)
```
