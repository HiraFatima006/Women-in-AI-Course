```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set_palette('pastel')
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data=pd.read_csv("startups.csv")
```

```python
# Dropping the Profit column from the 'x' data which consists of the the rest four columns
# 0-R&D 1-Admin 2-MS 3-State
```

```python
x=data.iloc[:,:-1].values
```

```python
y=data.iloc[:,4].values
```

```python
print(x,y)
```

```python
# Assigning a label to the State values
# 0-California, 1-Florida 2-New York
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
Label=LabelEncoder()
x[:,3]=Label.fit_transform(x[:,3])
print(x)
```

```python
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer([('encoder',OneHotEncoder(),[3])], remainder='passthrough')
x=np.array(ct.fit_transform(x))
print(x)                     
```

```python
print(pd.DataFrame(x).head())
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=14)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
print('R Squared', metrics.r2_score(y_test,y_pred))
```

```python
print('Mean Squared Error=', metrics.mean_squared_error(y_test,y_pred))
```

```python
print('Mean Absolute Error=', metrics.mean_absolute_error(y_test,y_pred))
```

```python
rmse=np.sqrt(metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Squared Error=', rmse)
```

```python
y_final_pred=model.predict(x)
```

```python
result=pd.concat([pd.DataFrame(y_pred,columns=['Prediction']),
                  pd.DataFrame(y_test,columns=["Test"])],axis=1)
result['Difference']=result['Prediction']-result['Test'] 
result
```

```python
y_fpred=pd.DataFrame(y_final_pred,columns=['Prediction'])
result=pd.concat([data,y_fpred],axis=1)
result['Difference']=result['Profit']-result['Prediction']
result
```

```python
metrics.r2_score(data['Profit'],y_final_pred)
```
